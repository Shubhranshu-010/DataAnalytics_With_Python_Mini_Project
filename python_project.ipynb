```python
#kpi analysis
#total sale
totalsales=df['TotalAmount'].sum()
print(totalsales)
```

```python
#total ordrs
totalorders=df['OrderID'].count()
print(totalorders)
```

```python
#avg sales
df['TotalAmount'].mean()
```

```python
#max quantity
df['Quantity'].max()
```

```python
#revenue by state
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False).head()
```

```python
#rev by category
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False).head(10)
```

```python
#top5 customer
df.groupby('CustomerID')['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
#order by payment mode
df.groupby('PaymentMethod')['OrderID'].count().sort_values(ascending=False).head(5)
```

```python
#total quantity sold for each product
df.groupby('ProductName')['Quantity'].sum()
```

```python
# sales trend by month and year

# Convert Order Date to datetime and create a Year-Month column
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year_Month'] = df['OrderDate'].dt.to_period('M')

# Group by the month and sum the Sales
monthly_sales = df.groupby('Year_Month')['TotalAmount'].sum().reset_index()

# Plot the trend
monthly_sales.plot(x='Year_Month', y='TotalAmount', kind='line', marker='o')

```

```python


# Group by State and Category, then sum the Sales
sales_by_state_category = df.groupby(['State', 'Category'])['TotalAmount'].sum().reset_index()

# Display the result
print(sales_by_state_category)
```

```python
#visualization
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.title("Rvenue by sale")
plt.xlabel('State')
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby('State')["TotalAmount"].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
df.groupby(['CustomerID','CustomerName'])['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
#category wise sales
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='pie')
plt.title('Category wise sales')
plt.show()
```

```python
df.groupby('ProductName')['TotalAmount'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 products by sales')
plt.xlabel('Product Name')
plt.ylabel('Total Amount')
plt.show()
```

```python
df.groupby('City')['TotalAmount'].sum().sort_values(ascending=True).head(10).plot(kind='bar')
plt.title('Top 10 cities by sales')
plt.xlabel('City')
plt.ylabel("Total Amount")
plt.show()
```
